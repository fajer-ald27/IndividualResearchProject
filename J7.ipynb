# **Clustering**

## *Clustering Pre-Processing*

Converting Date and Time to datetime then to numbers so clustering algorithm can be performed

```python
data_orig.head()
```

```python
data_copy=data_orig.drop('t',axis=1)
```

```python
data_orig.info()
```

```python
data_copy['date']=pd.to_datetime(data_copy['date'])
```

```python
data_copy['time']=pd.to_datetime(data_copy['time'])
```

```python
data_copy.info()
```

```python
data_copy['month']=data_copy['date'].dt.month
```

```python
data_copy['year']=data_copy['date'].dt.year
```

```python
data_copy['date']=data_copy['date'].dt.date
```

```python
data_copy['hour']=data_copy['time'].dt.hour
```

```python
data_copy['minute']=data_copy['time'].dt.minute
```

```python
data_copy['second']=data_copy['time'].dt.second
```

```python
data_copy.info()
```

```python
data_copy=data_copy.drop(['date','time'],axis=1)
```

```python
data_copy.head(70)
```

## *K-Means Clustering*

```python
from sklearn.cluster import KMeans
```

```python

# step 1
sse = {} 

# step 2 - iterate for a range of Ks and fit the scaled data to the algorithm. Use inertia attribute from the clustering object and 
# store the inertia value for that k 
for k in range(1, 15):
    kmeans = KMeans(n_clusters=k, max_iter=1000, random_state=1).fit(data_copy)
    sse[k] = kmeans.inertia_

# step 3
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()), 'bx-')
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```

```python
data_copy_labels = data_copy.copy(deep=True)
```

```python
n = 2

kmeans = KMeans(n_clusters = n, max_iter = 2000, random_state=1) #Apply the K-Means algorithm
kmeans.fit(data_copy) #Fit the kmeans function on the scaled data

#Adding predicted labels to the original data and scaled data 

labels = kmeans.predict(data_copy)
data_copy_labels['Labels'] = labels #Save the predictions on the scaled data from K-Means
data_orig['Labels'] = labels #Save the predictions on the scaled data from K-Means
```

```python
data_copy_labels['min5_effective_dl_cell_throughput']
```

```python
sns.scatterplot(x = 'min5_effective_dl_cell_throughput' , y = 'min5_effective_ul_cell_throughput' , data = data_copy_labels, hue='Labels')
```
