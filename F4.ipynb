```python
pdcp_data_volume.head()
```

```python
cells=pdcp_data_volume['cell name'].unique()
```

```python
pdcp_data_volume = pdcp_data_volume[pdcp_data_volume['pdcp_user_data_volume_dl_all_qci'].str.isnumeric()]
```

```python
pdcp_data_volume['pdcp_user_data_volume_dl_all_qci'].astype(float)
```

```python
pdcp_data_volume['pdcp_user_data_volume_dl_all_qci'] = pdcp_data_volume['pdcp_user_data_volume_dl_all_qci'].astype(float)
```

```python
for cell in cells:
  data = pdcp_data_volume[pdcp_data_volume['cell name'] == cell]
  try:
    plt.figure(figsize = (8,6))
    plt.plot(data['pdcp_user_data_volume_dl_all_qci'],label = cell,marker = 'o',markerfacecolor = 'red')
    plt.legend()
    plt.show()
  except:
    pass
  
```

```python
import seaborn as sns
for cell in cells:
  data = pdcp_data_volume[pdcp_data_volume['cell name'] == cell]
  try:
    sns.violinplot(data['pdcp_user_data_volume_dl_all_qci'])
    plt.show()
  except:
    pass
```
