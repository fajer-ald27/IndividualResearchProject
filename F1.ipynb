```python
label=[]
for index,row in x.iterrows():
  DL=row['min5_effective_dl_cell_throughput_per_qci']
  UL=row['min5_effective_ul_cell_throughput_per_qci']
  if (DL==0) and (UL==0):
    print(DL,UL)
    label.append(1)
  else:
    label.append(-1)
```

```python
row
```

```python
index
```

```python
x['anomaly']=label
```

```python
x.head()
```

```python
x.isnull().sum()
```

```python
y=x.copy()
```

```python
y=y.dropna()
```

```python
y.isnull().sum()
```

```python
y.head()
```

```python
y_features=y.drop(['starttime',	'stoptime','ecgi','anomaly'],axis=1)
```

```python
y_features.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(y_features)
```

```python
y_features_scaled = scaler.transform(y_features)
```

```python
print(y_features_scaled)
```

```python
k=pd.DataFrame(y_features_scaled , columns = y_features.columns)
```

```python
k.head()
```

```python
#k['anomaly']=y['anomaly']
#k=k.drop('anomaly',axis=1)
```

```python
k.head()
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf = IsolationForest(random_state=0).fit(k)
```

```python
clf.predict(k)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(k , y['anomaly'],test_size=0.33, random_state=42)
```

```python
clf = IsolationForest(random_state=0).fit(X_train,y_train)
```

```python
clf.predict(X_train)
```

```python
clf.predict(X_test)
```

```python
y_train_pred=clf.predict(X_train)
```

```python
y_test_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_train, y_train_pred)
```

```python
accuracy_score(y_test, y_test_pred)
```
