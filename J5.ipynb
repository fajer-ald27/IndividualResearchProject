```python
import pandas as pd
```

```python
import numpy as np
```

```python
data_orig = pd.read_csv('F_pdcp_data_volume.csv')
```

```python
data_orig.head()
```

```python
x=data_orig[['ecgi','starttime','pdcp_user_data_volume_dl_all_qci','pdcp_user_data_volume_ul_all_qci']]
```

```python
x.head(10)
```

```python
x.sort_values('starttime')
```

```python
x = x.apply(lambda x: x.str.strip()).replace('-', np.nan)
```

```python

x= x.dropna()
```

```python
x.to_csv('pdcp_data_volume_new.csv',index=False)
```

```python
test = x.copy()
```

```python
test.head(10)
```

```python
test['starttime'] = test['starttime'].astype('str')
```

```python
test.head(10)
```

```python
st=[]
for i in test['starttime'].values:
  i=str(i)
  i=i.replace('@',' ')
  st.append(i)

```

```python
st
```

```python
test['starttime']=st
```

```python
test.head()
```

```python
test['starttime']=pd.to_datetime(test['starttime'])
```

```python
test['starttime'].dt.date
```

```python
test['date']=test['starttime'].dt.date
```

```python
test.head()
```

```python
test['starttime'].dt.time
```

```python
test['time']=test['starttime'].dt.time
```

```python
test.head()
```

```python
test=test.drop('starttime',axis=1)
```

```python
test.head()
```

```python
test.info()
```

```python
ecgis = test['ecgi'].unique()
```

```python
print(ecgis)
```

```python
for ecgi in ecgis:
    df = test[test['ecgi'] == ecgi]
    df.to_csv(ecgi+'.csv', index = False)
```
